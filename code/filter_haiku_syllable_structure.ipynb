```python
import pandas as pd
import nltk
nltk.download('cmudict')
from nltk.corpus import cmudict
```

```python
def num_syllables_word(word):
    try:
        # from: https://datascience.stackexchange.com/questions/23376/how-to-get-the-number-of-syllables-in-a-word
        return [len(list(y for y in x if y[-1].isdigit())) for x in cmu_dict[word.lower()]][0]
    except:
        # print('error, word {} not in cmudict'.format(word))
        return 0

def has_correct_syllables(haiku):
    for i, sentence in enumerate(haiku):
        num_syllabes = 0
        try:
            for word in sentence.split():
                num_syllabes += num_syllables_word(word)
            if i == 0 or i == 2:
                if num_syllabes != 5:
                    return False
            elif i == 1:
                if num_syllabes != 7:
                    return False
                    break
        except:
            return False
    return True
```

```python
cmu_dict = cmudict.dict()
filepath = '../data/'
dataset = 'kaggle-jhalini-all_haiku.csv'

df = pd.read_csv(filepath+dataset)
correct_haikus =  df[[has_correct_syllables(haiku) for haiku in df[['0', '1', '2']].to_numpy()]]

num_correct = correct_haikus.shape[0]
print('Total haikus: {}'.format(df.shape[0]))
print('Correct Haikus: {} ({}%)\n'.format(num_correct, round(num_correct / df.shape[0] * 100, 2)))

correct_haikus.to_csv(filepath+'correct_haikus.csv')
correct_haikus.head(10)
```
