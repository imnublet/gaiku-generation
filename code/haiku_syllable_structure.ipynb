```python
import pandas as pd
import nltk
nltk.download('cmudict')
from nltk.corpus import cmudict
```

```python
def num_syllables(word):
    try:
        # from https://datascience.stackexchange.com/questions/23376/how-to-get-the-number-of-syllables-in-a-word
        return [len(list(y for y in x if y[-1].isdigit())) for x in cmu_dict[word.lower()]][0]
    except:
        # print('error, word {w} not in cmudict'.format(w=word))
        return 0

def get_correct_haikus(all_haikus):
    correct_haikus = []

    for haiku in all_haikus:
        correct_structure = True

        for i, sentence in enumerate(haiku):
            num_syllabes = 0
            try:
                for word in sentence.split():
                    num_syllabes += num_syllables(word)
                
                # print('num syllables for sentence {i}: {n}'.format(i=i, n=num_syllabes))
                if i == 0 or i == 2:
                    if num_syllabes != 5:
                        correct_structure = False
                elif i == 1:
                    if num_syllabes != 7:
                        correct_structure = False
            except:
                continue
                # print('error for sentence {i}: {s}'.format(i=i, s=sentence))
        
        if correct_structure:
            correct_haikus.append(haiku)

    return correct_haikus
```

```python
cmu_dict = cmudict.dict()
filepath = '../data/'
dataset = 'kaggle-jhalini-all_haiku.csv'

df = pd.read_csv(filepath+dataset, delimiter=',')
all_haikus = [[df['0'][i], df['1'][i], df['2'][i]] for i in range(df.shape[0])]

print('total haikus: {}'.format(len(all_haikus)))

correct_haikus = get_correct_haikus(all_haikus)
percentage_correct = round(len(correct_haikus) / len(all_haikus) * 100, 2)

print('correct haikus: {} ({}%)\n'.format(len(correct_haikus), percentage_correct))

print('first 10 haikus, seperated by line:')
for haiku in correct_haikus[:10]:
    print(haiku)
```
